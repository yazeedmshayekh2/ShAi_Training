```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df=pd.read_csv("company_sales_data.csv")
df
```

#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
####################################### Line Plot using Matplotlib visualization #########################################
line_plt = plt.plot(df["month_number"], df["total_profit"])
plt.ylim([100000, 500000])
plt.yticks(np.arange(100000, 500001, 100000))
plt.xticks(range(1,13))
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.title("Company profit per month using matplotlib");
```

```python
####################################### Line Plot using Seaborn visualization #########################################
line_sns = sns.lineplot(x= df['month_number'], y= df['total_profit'])
line_sns.set_ylim([100000, 500000])
line_sns.set_xticks(range(1,13))
line_sns.set_yticks(range(100000, 500001, 100000))
line_sns.set_title("Company profit per month using seaborn");
```

```python
#################################### Line Plot using Pandas visualization #####################################
line_pd = df.plot.line(x= 'month_number', y= 'total_profit', legend=None, ylim=([100000, 500000]))
line_pd.set_yticks(np.arange(100000, 500001, 100000))
line_pd.set_xticks(range(1,13))
line_pd.set_title("Company profit per month using pandas");
```

```python
############################## multiline plots using Matplotlib visualization #########################################
plt.plot(df['month_number'], df["facecream"], marker='o', linewidth=3, markersize=9)
plt.plot(df['month_number'], df["facewash"], marker='o', linewidth=3, markersize=9)
plt.plot(df['month_number'], df["toothpaste"], marker='o', linewidth=3, markersize=9)
plt.plot(df['month_number'], df["bathingsoap"], marker='o', linewidth=3, markersize=9)
plt.plot(df['month_number'], df["shampoo"], marker='o', linewidth=3, markersize=9)
plt.plot(df['month_number'], df["moisturizer"], marker='o', linewidth=3, markersize=9)
plt.legend(["Face cream Sales Data", "Face Wash Sales Data", "Toothpaste Sales Data", "Bathingsoap Sales Data", "Shampoo Sales Data", "Moisturizer Sales Data"])
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xticks(range(1,13))
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Sales data (using Matplotli)")
```

#### 3. Read face cream and facewash product sales data and show it using the bar chart.


- The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.  
- **Note:**It must be implemented in three methods  

The bar chart should look like this.


```python
#################################### bar Plot using Matplotlib visualization ####################################
x = np.arange(len(range(12)))  # the label locations
width = 0.3  # the width of the bars

plt.bar(x - width/2, df["facecream"], width)
plt.bar(x + width/2, df["facewash"], width)
plt.grid(linestyle='--', linewidth=2)

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.xlabel("Month number")
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.xticks([r for r in range(12)], range(1,13))

plt.legend(["Face cream Sales Data", "Face Wash Sales Data"], loc="upper left")

plt.show();
```

```python
####################################### bar Plot using Seaborn visualization #########################################
fig = plt.figure(figsize=(7,7))
sns.barplot(data= df, x= "month_number", y= "facecream", color="tab:blue")
sns.barplot(data= df, x= "month_number", y= "facewash", color="orange")
plt.grid(linestyle='--', linewidth=2)
plt.legend(["Face cream Sales Data", "Face Wash Sales Data"], loc="upper left");
```

```python
#################################### bar Plot using Pandas visualization ###################################
barplot_pd =df[["month_number", "facecream", "facewash"]].set_index('month_number').plot(kind='bar', width=0.8, grid=True)
plt.grid(linestyle='--', linewidth=2)
plt.xticks(rotation=0)
barplot_pd.set_title("facewash and facecream sales data(using pandas)")
barplot_pd.set_ylabel("Sales Units in number")
barplot_pd.set_xlabel("Month Number")
barplot_pd.legend(["Face cream Sales Data", "Face Wash Sales Data"], loc="upper left");
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
sub_df= df[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]]
labels=["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]
slices= []
for sli_ce in sub_df:
    slices.append(sub_df[sli_ce].sum())
slices
```

```python
slices_df= pd.DataFrame(slices, index=labels)
slices_df["labels"]= slices_df[0]
del(slices_df[0])
slices_df
```

```python
|#################################### pie Plot using Matplotlib visualization #####################################
plt.pie(slices, labels=labels, autopct='%1.1f%%')
plt.legend(labels, loc="lower right");
```

```python
#################################### pie Plot using Pandas visualization #####################################
slices_df.plot.pie(y='labels', title=None, autopct='%1.1f%%')
plt.legend(labels, loc="lower right")
plt.ylabel("")
```
